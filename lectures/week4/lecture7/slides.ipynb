### Yule's regression equation

$$\Delta\text{Paup} = a + b\times\Delta\text{Out} + c\times\Delta\text{Old} + d\times\Delta\text{Pop} + \text{error}$$

* $\Delta$ is percentage change over time
* $\text{Paup}$ is percentage of paupers 
* $\text{Out}$ is the out-relief ratio  
  $(\text{number receiving assistance outside the poor-house}) / \text{number inside})$
* $\text{Old}$ is the percentage of the population over 65
* $\text{Old}$ is the population

* Data from about 600 unions, obtained from censuses of 1871, 1881, 1891
* Equations fit separately for rural, mixed, urban, metropolitan

### Example results (metropolitan)

* 1871--1881:
$$\Delta\text{Paup} = 13.19 + 0.755\times\Delta\text{Out} - 0.022\times\Delta\text{Old} - 0.3222\times\Delta\text{Pop} + \text{error}$$

* 1881--1891:
$$\Delta\text{Paup} = 1.36 + 0.324\times\Delta\text{Out} + 1.37\times\Delta\text{Old} - 0.369\times\Delta\text{Pop} + \text{error}$$

## Path models and response schedules

- Path models are a way to formalize a set of relationships/regressions between different r.v.'s.
- "Path models are often held out as rigorous statistical engines for inferring causation from association."
- Depends strongly on the assumptions you make of course!

This model encompasses two regression equations:
1. $Y = a + b X + \epsilon$
2. $Z = c + d X + e Y + \gamma$

We want to argue that if $X$ increases by one unit **holding $Y$ fixed** then $Z$ goes up by $d$ units.

To do this we have to think more carefully about our data generating model.

\begin{align}
Y_{i,x} &=a+bx+\delta\\
Z_{i,x,y}&=c+dx+ey+\epsilon_i
\end{align}

Assumptions:
1. $\delta_i \perp \epsilon_i$
2. $\delta_i \perp \delta_j$ and $\epsilon_i\perp\epsilon_j$ for all $i, j$.
3. $\delta_i$ and $\epsilon_i$ have the same distribution across all subjects (but not necessarily the same distribution.)
4. $\mathbb{E}\delta_i=0$ and $\operatorname{var}\delta_i<\infty$ and similarly for $\epsilon$.
5. $X_i \perp \delta_i,\epsilon_i$ (exogeneity).

> Inferences get made by ignoring the iffiness of the assumptions.

### Potential outcomes
- In the causal inference literature $Y_{i,x}$ and $Z_{i,x,y}$ are called **potential outcomes**.
- $Y_{i,x}$ is what we would observe for subject $i$ if we set $x$ by intervention.
- **We only ever observe one setting of $x$ for each subject**.

### Observational Studies

* In an observational study, the researcher **does not** decide who will get treatment and who will get control.

* (What would that look like for smoking?)

* Often the subjects decide themselves.

* Sometimes some other external factor decides.

### Confounding

* Key problem with observational studies: Confounding.  
* Confounder: A variable that (1) is associated with treatment, and (2) has an effect on response.  
* When confounders are present, it's hard to say whether any observed difference in response between the treatment group and the control group is due to treatment, or due to the confounders.

### Example: Smoking

* Smokers have a higher rate of lung cancer.  
* Does smoking cause lung cancer?  
* For a long time, people were unsure, because of possible confounders.

### Possible confounders
* Age  
* Gender  
* Occupation  
* Urban / rural  
* Alcohol  
* Genetics  
* And many more...

### Cross Tabulation
* To deal with confounding, a researcher might make more and more refined comparisons.

* Example: Compare men who smoke to men who don't smoke; compare women who smoke to women who don't smoke.

* Compare men who live in cities who smoke to men who live in cities who don't smoke. 
    - Also compare men who live in the countryside who smoke to men who live in the countryside who don't smoke.  
    - And similarly for women.

* This process is called *cross tabulation.*  (Often the results are presented in a table.)

### Problems with cross-tabulation
There are two main problems with cross tabulation:

* Very quickly run out of data if you look at several confounders (curse of dimensionality)

* Always possible you forgot an important confounder

The first problem, in principle, could be solved by more data. However the second cannot.

### Example: Smoking

- We can easily calculate $\mathbb{P}(\text{cancer} \mid \text{smoker})$ and $\mathbb{P}(\text{cancer} \mid \text{nonsmoker})$.

- But the question we're really interested in is 

$$\mathbb{P}(\text{cancer} \mid \textbf{do}(\text{nonsmoker}\to\text{smoker})) - \mathbb{P}(\text{cancer} \mid \textbf{smoker})$$.

That's the causal effect of smoking on cancer.

- This is called the *counterfactual* or *potential outcome*.

- We never observe it! We only observe what actually happened.

## Design-based Inference
- So far we have seen examples of *model-based* inference
    - Assume a specific model for the data
    - Make inferences about the model

- How can this go wrong?

- An alternative to making assumptions about the data is to make assumptions about the sampling mechanism.

- This leads to model-based inference.

### Example: Electricity and Plants
* Question: Do small electric currents in the soil alter the growth of plants?

- (Note that this is a causal claim.)

* Experiment: 
    * Grow 8 plants from seed
    * Subject 4 to small electric currents
    * Measure height of plants after 1 month

### How to Design the Experiment? 

Plenty of details to consider, but let's agree on the following:
* One seed per pot
* Randomize seeds to pots (good -- but not critical, and **not** sufficient)
* Randomize treatment (electricity) to pots
* Ideally: Keep treatment setup as similar for each pot as possible
    * have all the wiring in place, a switch some distance away
    * similar in spirit to having a placebo

## Other examples

### Vitamins and Kangaroos

It is claimed that a vitamin supplement helps kangaroos learn to run a special maze with high walls.  To test whether this is true, 20 kangaroos are divided up into 10 pairs.  In each pair, one kangaroo is selected at random to receive the vitamin supplement; the other is fed a normal diet.  The kangaroos are then timed as they learn to run the maze.  In 9 of the 10 pairs, the treated kangaroo learns to run the maze more quickly than its untreated partner.

### Drug trial

A drug is thought to reduce the risk of heart attacks.  To test the drug an experiment is conducted on 18 patients at high risk of a heart attack.  Nine patients are randomly assigned to take the drug, while the other nine receive a placebo.  The patients are then followed for three years.  At the end of the experiment, four of the patients in the placebo group had had a heart attack, while only one patient in the treatment group had had a heart attack.

## Nonparametric testing

- In each of the preceding examples we are testing

$$H_0: F=G\quad\text{versus}\quad H_1:F\neq G$$

where $F,G$ are the data generating distributions of two different groups.

- (Plants): $F$=height without electricity; $G$=height with electricity.

- (Tea): $F$=taste when milk added first; $G$=taste when milk added second.

## Permutation testing
- Under $H_0$, the data are *exchangeable*:

$$\mathbb{P}(\underbrace{X_1,\dots,X_n}_{\text{group 1}},
\underbrace{Y_1,\dots,Y_m}_{\text{group 2}})
=\mathbb{P}(\underbrace{Z_{\sigma(1)},\dots,Z_{\sigma(n +m )}}_\text{any permutation})$$

- Randomness is in the assignment of observations to groups.

- The $p$-value for any stat $T$ is simply:

$$\frac{1}{(n+m)!} \sum_\sigma \mathbf{1}\{T(X_1,\dots,X_n,Y_1,\dots,Y_m) < T(Z_{\sigma(1)},\dots,Z_{\sigma(n+m)})\}$$

## Permutation testing for the plants example

```r
# Data:
library(tidyverse)
df <- tibble(
    x = c(  0,   1,   0,   0,   1,   1,   1,   0),
    y = c(8.6, 8.4, 8.9, 8.6, 8.5, 8.1, 8.3, 8.7)
)
```

```r
# observed mean
mean.diff <- function(df) {
    df %>% group_by(x) %>% summarize(mean(y)) %>% pluck(2) %>% diff
}
mean.diff(df)
```

```r
# shuffled means
shuf.mean <- function(i) {
    set.seed(i)
    df %>% mutate(x = sample(x)) %>% mean.diff
}

vals <- 1:1000 %>% map_dbl(shuf.mean) %>% abs 
```

```r
summary(vals)
```
