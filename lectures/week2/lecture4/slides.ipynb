## Advice for all kinds of talks
(Lots of opinions on this. Here are mine.)

1. Know your audience.
2. Have a purpose.
3. Less is more.
4. Tell a story.
5. Practice!

### Know your audience
- What is their background?

- Why have they come to see you speak?
    - What are they hoping to get out of it?
    - What can you do to make them feel like their time is well-spent? 
    - Put yourself in their shoes: would you enjoy attending your own talk? If not, change the talk!

![Reading level of major news outlets](https://readable.com/wp-content/uploads/2019/04/transparent-graph-w-spell-correction.001.png)

-  Non-technical talks: the audience is there to learn. 
-  **Common mistake in academic talks**: assuming the audience understands your research.
    - This is not as common as you hope/think.
    - Even people "in your area" can get lost if you move too fast or your slides are unclear.
    - It's better to go too slow that too fast.

### Have a purpose

- What is your goal as a speaker?
    - Advertise your work?
    - Educate the audience?
    - Get a job?

- Does each element of your presentation serve this purpose? If not, cut.

### Less is more

- There is a limit to the pace at which the human mind can assimilate information.

- Modern presentation aids (slides) makes it easy to **greatly** exceed this limit. Don't fall into this trap!
    - Err on the side of fewer slides, less text.
    - Instead of presenting lots of ideas/results, focus on presenting a few of them well.
    - A picture really is worth 1000 words in many cases.

- Do not go over time unless asked.
   - I have never attended a talk where people got upset because the speaker ended a few minutes early.
   - Presenting slides takes longer than you think.
   - For a 50m talk I budget ~35 slides.

### Tell a story

- Your talk should have a beginning, middle, and end.

- Most stats talks can be organized as follows:
    - State the problem and explain why it is important.
    - Here is what we did to solve it.
    - Here are the results.
    - Future work/what remains to be done.

### Practice!

- Practice your talk in front of friends. (Easy to do in grad school.)

- Practice in front of the mirror.

-  Record yourself (and watch it afterwards).

## Tricks of the trade

- Overcoming nerves:
    - The opening minutes are usually the hardest part.
        - Open with a joke.
        - Memorize the first few slides.
    - Memorize what your are going to say (but not necessarily how you are going to say it).

- Presenter notes, used sparingly, can be a great addition. (But don't just read from them.)

- Anticipate questions, have backup slides ready.

- Audience participation/interaction is great for breaking the ice.

- There is no substitute for practice.
